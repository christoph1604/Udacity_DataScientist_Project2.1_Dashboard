```python
import pandas as pd
```

```python
df = pd.read_csv("./covid19-spain-cases.csv")
df=df[["ccaa", "fecha", "casos", "muertes", "nuevos"]]
df.columns=["Region", "Date", "Cases", "Dead", "New_Cases"]

df["Cases"]=df["Cases"].fillna(0)
df["Dead"]=df["Dead"].fillna(0)
df["New_Cases"]=df["New_Cases"].fillna(0)

df["Cases"]=df["Cases"].astype(int)
df["Dead"]=df["Dead"].astype(int)
df["New_Cases"]=df["New_Cases"].astype(int)
df["Date"]=pd.to_datetime(df["Date"], format='%Y-%m-%d', errors='ignore')

df

```

```python
df.to_csv('covid19-spain-cases.csv', index=False)
```
